```python
import spacy

# Load the installed model
nlp = spacy.load('en_core_web_sm')

# Define your file paths
file_paths = [
    "C:\\Users\\User\\SDA250Mywork\\A2_data\\5c5e50711e67d78e27616b23.txt",
    "C:\\Users\\User\\SDA250Mywork\\A2_data\\5c5da7aa1e67d78e275f8a3c.txt",
    "C:\\Users\\User\\SDA250Mywork\\A2_data\\5c5d532a795bd2d5c282a094.txt",
    "C:\\Users\\User\\SDA250Mywork\\A2_data\\5c5d3f0a1e67d78e275e5788.txt",
    "C:\\Users\\User\\SDA250Mywork\\A2_data\\5c5d3e251e67d78e275e54b5.txt"
]

# Function to read file content
def read_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

# Process and print entities recognized as people in each file
for file_path in file_paths:
    content = read_file(file_path)
    doc = nlp(content)
    print(f"Entities recognized as people in {file_path}:")
    for ent in doc.ents:
        if ent.label_ == "PERSON":
            print(ent.text)
    print("\n---\n")
```
